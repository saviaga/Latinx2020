## Community  Group Posts

```python
#path = "Latinx2020/"
posts_community = pd.read_csv("FB_groups_States_immigration.csv")
#We rename the colum Page Name to be able to concat with Group Posts
posts_community.rename(columns={'Page Name':'Group Name'}, inplace=True)
posts_community.shape
```

```python
posts_community["Created"].head(5)
```

```python
posts_community['converted_date'] = pd.to_datetime(posts_community["Created"])
posts_community['converted_date'].head(3)
```

```python
posts_community['converted_date']=posts_community['converted_date'].dt.date
```

```python
posts_community.shape
```

```python
group_date_community = posts_community.groupby('converted_date').agg({'Facebook Id':'count'}).reset_index().rename(columns={'Facebook Id':'count'})
group_date_community
```

```python
group_date_community['category'] = 'community'
```

## News Outlets Posts 

```python
#path = "Latinx2020/"
posts_news = pd.read_csv("FB_groups_News_Media_Outlets_immigration.csv",parse_dates=['Created'])
#We rename the colum Page Name to be able to concat with Group Posts
posts_news.rename(columns={'Page Name':'Group Name'}, inplace=True)
posts_news.shape
```

```python
posts_news['converted_date'] = pd.to_datetime(posts_news["Created"])
posts_news['converted_date'].head(3)
```

```python
posts_news['converted_date'] = posts_news['converted_date'].dt.date
posts_news['converted_date'].head(3)
```

```python
posts_news
```

```python

group_date_news = posts_news.groupby('converted_date').agg({'Facebook Id':'count'}).reset_index().rename(columns={'Facebook Id':'count'})
group_date_news
```

```python
group_date_news['category'] = 'news_outlets'
```

## Political Groups Posts

```python
#path = "Latinx2020/"
posts_pol = pd.read_csv("Topics_FB_Political_Posts_per_partyimmigration.csv")
#We rename the colum Page Name to be able to concat with Group Posts
posts_pol.rename(columns={'Page Name':'Group Name'}, inplace=True)
posts_pol.shape
```

```python
posts_pol['converted_date'] = pd.to_datetime(posts_pol["Created"])
posts_pol['converted_date'].head(3)
```

```python
posts_pol['converted_date'] = posts_pol['converted_date'].dt.date
```

```python
posts_pol.shape
```

```python
group_pol = posts_pol.groupby('converted_date').agg({'Facebook Id':'count'}).reset_index().rename(columns={'Facebook Id':'count'})
group_pol
```

```python
group_pol['category'] = 'Political_Groups'
```

```python
all_groups = pd.concat([group_date_news,group_date_community,group_pol],axis=0)
all_groups.sort_values('converted_date',ascending=True)
```

## Logarithmic Scale

```python

fig = plt.gcf()
fig.set_size_inches( 16, 10)
g_results=sns.lineplot(data=all_groups, x='converted_date', y='count',hue='category')
g_results.set(yscale='log')

plt.xticks(rotation=70)
plt.tight_layout()
```

## Normal Scale

```python

fig = plt.gcf()
fig.set_size_inches( 16, 10)
g_results=sns.lineplot(data=all_groups, x='converted_date', y='count',hue='category')

plt.xticks(rotation=70)
plt.tight_layout()
```
